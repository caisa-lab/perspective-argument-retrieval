### 1. Separate demographic data in Arguments arguments to individual columns

```python
keys = list(arguments.demographic_profile[0].keys())
```

```python
def getValues(data, key):
    return data[key]
```

```python
for key in keys:
    arguments[key] = arguments.demographic_profile.apply(getValues, args=(key,))
```
